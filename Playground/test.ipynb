```python
import xml.etree.ElementTree as ET
from xml.etree.ElementTree import ParseError
import pandas as pd
from glob import glob
import os.path as op
import numpy as np
from collections import Counter
import operator
from string import punctuation
from nltk import SnowballStemmer
from nltk import pos_tag
import nltk
```

```python
def add_records(path, all_records):
    xml_data = open(path, encoding="utf8").read()
    xml_data = "<problems>" + xml_data + "</problems>"
    problems = ET.XML(xml_data)
    
    for problem in problems:
        for child in problem:
            record = {}
            for grandChildren in child:
                if grandChildren.tag == "choices":
                    for choice in grandChildren:
                        if str(choice.get('id'))!= None:
                            record[choice.tag + '_' + str(choice.get('id'))] = choice.text
                else:
                    record[grandChildren.tag] = grandChildren.text
                                
                        
            all_records.append(record)
```

```python
def add_year(path):
    tree = ET.parse(path)  
    root = tree.getroot()
    
    root.remove(root[0])
    
    for elem in root:  
        if elem.tag== "problem":
            
            #Add the tag "year"
            attrib = {}  
            year = elem.makeelement('year', attrib)  
            year.text = path[27:31]
            elem.append(year)
            
            #Add the tag "topic"
            question = elem.find('question').text
            num= int(question.split(".")[0])
            attrib = {}  
            topic = elem.makeelement('topic', attrib)  
            if num in range (1, 19):
                topic.text = 'Ethical and Professional Standards'
                elem.append(topic)
            
            elif num in range(19, 33):
                topic.text = 'Quantitative Methods'
                elem.append(topic)
                
            elif num in range(33, 45):
                topic.text = 'Economics'
                elem.append(topic)
                
            elif num in range(45, 69):
                topic.text = 'Financial Statement Analysis'
                elem.append(topic)
            
            elif num in range(69, 79):
                topic.text = 'Corporate Finance'
                elem.append(topic)
                
            elif num in range(79, 91):
                topic.text = 'Equity Investments'
                elem.append(topic)
                
            elif num in range(91, 97):
                topic.text = 'Derivative Investments'
                elem.append(topic)
                
            elif num in range(97, 109):
                topic.text = 'Fixed Income Investments'
                elem.append(topic)
                
                
            elif num in range(109, 115):
                topic.text = 'Alternative Investments'
                elem.append(topic)
                
            elif num in range(115, 121):
                topic.text = 'Portfolio Management'
                elem.append(topic)
            
                
    #Saving the result in a new path            
    tree.write('mia'+path[27:31]+'.xml')  
    return 'mia'+path[27:31]+'.xml'
```

```python
all_records = []
file_paths = sorted(glob(op.join('..','Data', 'qa_mock_exams', '*', '*.xml')))
for path in file_paths:
    year=int(path[27:31])
    if year in range (2009, 2010):
        new_file=add_year(path)
        try:
            add_records(new_file, all_records)
        except ParseError as e:
            print("Exception in file '{}': {}".format(path, e))

problems_df = pd.DataFrame(all_records)

#Pour eliminer les phrases "Select exactly 1 answer(s) from the following:"
#problems_df['question'] = problems_df['question'].map(lambda x: x.replace("Select exactly 1 answer(s) from the following:  \t", ""))
```
