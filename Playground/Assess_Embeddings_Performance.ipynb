```python
import json
import spacy
import numpy as np
```

```python
results = json.load(open('../Results/results_05_22_19_19.json', 'r'))

wrong_results = []

for cat in ['DEF_KEYWORD', 'DEF_KEYWORD_START_END', 'KEYWORD_DEF', 'KEYWORD_DEF_START_END']:
    for file, result in results[cat].items():
        if not result['random_answer'] and result['real_answer'] != result['predicted_answer']:
            supposed_index = ord(result['real_answer']) - ord('A')
            predicted_index = ord(result['predicted_answer']) - ord('A')
            supposed_choice = result['choices_results'][supposed_index]
            predicted_choice = result['choices_results'][predicted_index]
            if len(supposed_choice['scores']) > 0 and len(predicted_choice['scores']) > 0:
                wrong_results.append(result)
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
print(nlp.vocab.strings.add('if'))
print(nlp.vocab.strings.add('If'))
```

```python
problem = wrong_results[1]
question = nlp(problem['question'])
choice1 = nlp(list(problem['choices_results'][0]['scores'].keys())[0])
choice2 = nlp(list(problem['choices_results'][1]['scores'].keys())[0])
choice3 = nlp(list(problem['choices_results'][2]['scores'].keys())[0])

print(json.dumps(problem, indent=4))
print()
print(question.similarity(choice1))
print(question.similarity(choice2))
print(question.similarity(choice3))
```

```python
def print_oov(sentence):
    for token in sentence:
        if nlp.vocab[token.string.strip()].is_oov:
            print(token.string)
            
print_oov(choice1)
print_oov(choice2)
print_oov(choice3)
```
